# XGBoost important features

What features are the models using?

```python
# load all XGBoost models and the data they were trained on

# read in data
import pandas as pd
beta_array = pd.read_csv('data/training_testing_maxgap_50_newdata.csv')

import joblib
models = []
for i in range(4):
    model = joblib.load('figs_xgboost/xgboost_models/' + str(i) + '.pkl')
    models.append(model)
    
    
```

```python
models
```

```python
beta_array
```

```python
list(model.feature_importances_)[0:10]
```

```python
beta_array = beta_array.reset_index()
```

```python
beta_array
```

```python
i = 0

features_list = []
for i in range(4):
    model = models[i]

    features = beta_array.iloc[model.feature_importances_ > 0, ]

    features = features[['seqnames', 'start']]
    features_list.append(features)
```

```python
features_list
```

```python
# save features_list
for i in range(len(features_list)):
    features_list[i].to_csv('figs_xgboost/xgboost_models/' + str(i) + '_important_features.csv')
                         
```

```python
features_union = pd.concat(features_list).drop_duplicates()
features_union.to_csv('figs_xgboost/xgboost_models/features_union.csv', index = False)
```

```python
# save input features for use as background when analysing features
beta_array[['seqnames', 'start']].to_csv('figs_xgboost/xgboost_models/input_features_for_background.csv', index = False)
```

```python
# now lets look at the overlap
string_features = []
for i in range(4):
    string = features_list[i]['seqnames'].astype('str') + '_' + features_list[i]['start'].astype('str')
    string_features.append(list(string))
```

```python
[len(s) for s in string_features]
```

```python
from functools import reduce
all_4 = reduce(set.intersection, [set(l_) for l_ in string_features])
all_4
len(all_4)
```

```python
%matplotlib inline

from venn import venn
sets = {i:set(string_features[i]) for i in range(len(string_features))}
venn(sets, figsize=(10, 10))
```

```python
all_4
```

```python
# which of these all_4 are the most important?
import numpy as np

intersected_features = pd.DataFrame(all_4, columns=['feature_string'])
seqnames = [i[0] for i in intersected_features['feature_string'].str.split('_')]
start = [i[1] for i in intersected_features['feature_string'].str.split('_')]
intersected_features['seqnames'] = seqnames
intersected_features['start'] = start
intersected_features


for i in range(len(intersected_features)):
    print(i)
    seqname = intersected_features.iloc[i]['seqnames']
    s = int(intersected_features.iloc[i]['start'])
    index = beta_array[beta_array['seqnames'].astype('string') == seqname][beta_array['start'] == s].index.values[0]
    print(index)
    model_num = 0
    for m in models:
        importance = m.feature_importances_[index]
        intersected_features.loc[i, model_num] = importance
        model_num = model_num + 1
```

```python
intersected_features['mean_importance'] = intersected_features[[0, 1,2,3]].mean(axis = 1)
```

```python
intersected_features = intersected_features.sort_values(by = 'mean_importance', ascending=False)
```

```python
intersected_features[['seqnames', 'start', 'mean_importance']]
```

```python
# turn this into a bed file
bed = intersected_features[['seqnames', 'start', 'start', 'mean_importance']]
bed.columns = ['seqnames', 'start', 'end', 'mean_importance']
bed['start'] = bed['start'].astype(int) - 1 # bed files index from 0
bed.to_csv('figs_xgboost/xgboost_models/intersected_features.bed', header = False, index = False, sep = '\t')
```

```python
# save background features as bed file
bed_bg = beta_array[['seqnames', 'start', 'start']]
bed_bg.columns = ['seqnames', 'start', 'end']
bed_bg['start'] = bed_bg['start'].astype(int) - 1 # bed files index from 0

bed_bg['seqnames'] = 'chr' + bed_bg['seqnames'].astype(str)
bed_bg

bed_bg.to_csv('figs_xgboost/xgboost_models/input_features_for_background.bed', header = False, index = False, sep = '\t')
```

# Do these features overlap with other kidney cancer gene lists?

We have mapped these features to genes (see map_features_to_genes.Rmd). 
We also have a list of (differentially expressed) Kidney cancer genes from Renal Gene Expression Database (RGED): a relational database of gene expression profiles in kidney disease. http://rged.wall-eva.net/



```python
pd.read_csv('figs_xgboost/xgboost_models/'+str(i)+'_gene_list.csv', sep= ' ')
```

```python
# read in our found genes and save out the union of genes

gene_lists = pd.DataFrame()
for i in range(4):
    genes = pd.read_csv('figs_xgboost/xgboost_models/'+str(i)+'_gene_list.csv', sep= ' ')
    gene_lists = gene_lists.append(genes)

gene_lists = gene_lists.drop_duplicates()
gene_lists
```

```python
# save both gene_names and ensembl ids

gene_lists['gene_name'].dropna().drop_duplicates().to_csv('figs_xgboost/xgboost_models/features_mapped_to_genes.csv', index = False, header = False)
gene_lists['ensembl_gene_id'].dropna().drop_duplicates().to_csv('figs_xgboost/xgboost_models/features_mapped_to_genes_ensembl.csv', index = False, header = False)
```

```python
gene_lists = set(gene_lists['gene_name'].dropna())
len(gene_lists)

# read in RGED database
RGED = pd.read_csv('RGED_database_kidney_carcinoma_genes_diff_expression_microarrays.txt', header = None)
len(RGED)
```

```python
# look at overlap
overlap = gene_lists.intersection(RGED[0])
len(overlap)
```
