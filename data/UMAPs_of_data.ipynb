```python
diagnoses = pd.read_csv('training_testing_diagnoses_maxgap_50_newdata.csv')['overallpath'][2:]
diagnoses[diagnoses == 5] = 4 # pooling type 1 and 2 pRCC into one type

label_mapping = {0:'normal', 1:'ccRCC', 2:'chRCC', 3:'oncocytoma', 4:'pRCC', 5:'pRCCT2', 6:'other'}
```

```python
train_test_source = pd.read_csv('training_testing_diagnoses_maxgap_50_newdata.csv')['colname'][2:]
train_test_source = ['TCGA' if 'TCGA' in t else 'Chopra' if 'R' in t else 'Cambridge' for t in train_test_source]
```

```python
def get_features(model_num):
    features = pd.read_csv('../figs_xgboost/xgboost_models/'+str(model_num)+'_important_features.csv')
    features_to_take = list(features['Unnamed: 0'])
    return(features_to_take)

import umap
def fit_umap(met_array):
    reducer = umap.UMAP(n_components=2)
    embedding = reducer.fit_transform(met_array)
    return(embedding)

def plot_umap(embedding, labels, name, source):
    labels_mapped = [label_mapping[p] for p in labels]

    import seaborn
    import matplotlib.pyplot as plt
    plt.figure(figsize=(8, 8))
    # make 16 colour palette:
    new = seaborn.color_palette("Paired")[0:len(set(labels))]
    
    # colour scheme: c("Normal"="#56B4E9","KIRC"="#007d5b", "KICH"="#000000", "ONC"="#D55E00", "KIRP"="#CCBB44")
    palette = ["#56B4E9", "#007d5b", "#000000", "#D55E00", "#CCBB44"]
    
    if len(set(source)) == 1:
        markers = ['o']
    else:
        markers = {'TCGA':'x', 'Cambridge':'+', 'Chopra':'|', 'External Evelonn': '_'} 

    if len(set(labels)) == 5: # if we just have one pooled prcc type 
        ax = seaborn.scatterplot(x = embedding[:, 0], y = embedding[:, 1], hue = labels_mapped, style = source,  markers=markers, palette = palette, legend="full", s = 60, alpha = 1, edgecolor='#000000', hue_order = ['normal','ccRCC','chRCC','oncocytoma','pRCC']) 
        
    else:
        ax = seaborn.scatterplot(x = embedding[:, 0], y = embedding[:, 1], hue = labels_mapped, style = source, markers=markers, palette = palette, legend="full", s = 20, alpha = 1, edgecolor='#000000', linewidth = 0, hue_order = ['normal','ccRCC','chRCC','oncocytoma','pRCC','pRCCT2'])
    ax.set_xlabel('UMAP_1', fontsize=16)
    ax.set_ylabel('UMAP_2', fontsize=16)
    
    plt.legend(prop={'size': 16}, bbox_to_anchor=(1.1, 1))
    
    plt.savefig('UMAPs/' + name + '.svg', bbox_inches='tight')

    
```

```python

def plot_umap_with_highlight(embedding, labels, name, source, to_highlight, to_write = [], s = -1):
    label_mapping = {0:'normal', 1:'ccRCC', 2:'chRCC', 3:'oncocytoma', 4:'pRCC', 5:'pRCCT2', 6:'other'}
    labels_mapped = [label_mapping[p] for p in labels]

    import seaborn
    import matplotlib.pyplot as plt
    plt.figure(figsize=(8, 8))
    # make 16 colour palette:
    new = seaborn.color_palette("Paired")[0:len(set(labels))]
    
    # colour scheme: c("Normal"="#56B4E9","KIRC"="#007d5b", "KICH"="#000000", "ONC"="#D55E00", "KIRP"="#CCBB44")
    if 'stanford' in name:
        palette = ["#56B4E9", "#007d5b", "#000000", "#D55E00"]
        hue_order = ['normal','ccRCC','chRCC','oncocytoma']
    else:
        palette = ["#56B4E9", "#007d5b", "#000000", "#D55E00", "#CCBB44"]
        hue_order = ['normal','ccRCC','chRCC','oncocytoma','pRCC']
    
    
    if len(set(source)) == 1:
        markers = ['o']
        if s == -1: # only set s if not given a value
            s = 20
    else:
#         markers = ['o', 's', 'p']
        markers = {'TCGA':'x', 'Cambridge':'+', 'Chopra':'|'} 
        if s == -1: # only set s if not given a value
            s = 60
        
    if len(set(labels)) != 6: # if we just have one pooled prcc type 
        ax = seaborn.scatterplot(x = embedding[:, 0], y = embedding[:, 1], hue = labels_mapped, style = source,  markers=markers, palette = palette, legend="full", s = s, alpha = 1, edgecolor='#000000', hue_order = hue_order) 
    else:
        ax = seaborn.scatterplot(x = embedding[:, 0], y = embedding[:, 1], hue = labels_mapped, style = source, markers=markers, palette = palette, legend="full", s = s, alpha = 1, edgecolor='#000000', hue_order = ['normal','ccRCC','chRCC','oncocytoma','pRCC','pRCCT2'])
    ax.set_xlabel('UMAP_1', fontsize=16)
    ax.set_ylabel('UMAP_2', fontsize=16)
    
    plt.legend(prop={'size': 16}, bbox_to_anchor=(1.1, 1))
    
    
    if s == -1: # if s not specified, set smaller_s
        smaller_s=50
    else:
        smaller_s = s-10 # else make smaller_s a bit smaller than the given s
    plt.scatter(x = embedding[to_highlight, 0], y = embedding[to_highlight, 1], s=smaller_s, facecolors='none', edgecolors='r', linewidth=2)
    

    if to_write != []:
        for x, y, s in zip(embedding[:, 0], embedding[:, 1], to_write):
            if s != '':
                plt.text(x+0.01, y+0.01, s, fontsize=12)

    plt.savefig('UMAPs/' + name + '.svg', bbox_inches='tight')
    

```

### UMAP of train and test, using only XGBoost features

These seem to separate the classes well, for the majority of samples. The oncocytoma and chromophobe classes are close together, but there is separation for the majority of those samples.

Plots both with highlights of incorrect samples and without

```python
# get test predictions (from all models)
preds = np.loadtxt('../figs_xgboost/all_test_predictions')
is_correct = pd.DataFrame({'diagnosis': diagnoses, 'prediction': preds})
is_correct['correct'] = is_correct['prediction'] == is_correct['diagnosis']
is_correct
to_highlight = is_correct['correct'] == False


for model_num in range(4):
    features_to_take= get_features(model_num)
    data_xgbfeatures = beta_array.iloc[:, features_to_take]
    
    # preprocess data - remove NAs
    data_xgbfeatures = data_xgbfeatures.fillna(0)
    
    # fit UMAP
    embedding = fit_umap(data_xgbfeatures)
    
    # plot umap
    plot_umap(embedding, diagnoses, 'train_and_test_only_XGBoost_features_' + str(model_num), train_test_source)
    # and same UMAP but with highlight
    plot_umap_with_highlight(embedding, diagnoses, 'train_and_test_only_XGBoost_features_' + str(model_num) + '_with_highlight', train_test_source, to_highlight)
    
    
```

# UMAP of Evelonn using XGBoost features

```python
# read in and process the data:

data = pd.read_csv('Evelonn_external/preprocessed_validation_m_vals_from_normalised_data.csv', sep = ',')
data = data.iloc[:,3:]
Evelonn_diagnoses = pd.read_csv('Evelonn_external/preprocessed_validation_beta_diagnoses_from_normalised_data.csv')
mapping = {'Normal': 0, 'ccRCC': 1}
Evelonn_diagnoses['overallpath'] = [mapping[l] for l in Evelonn_diagnoses['diagnosis']]


# get into right shape
met_array = np.array(data).transpose()
met_array.shape # rows are samples, columns are features
met_array
```

```python

to_plot = met_array
to_plot = pd.DataFrame(to_plot)

for model_num in range(4):

    features_to_take= get_features(model_num)
    data_xgbfeatures = to_plot.iloc[:, features_to_take]
    data_xgbfeatures.shape
    
    # preprocess data - remove NAs
    data_xgbfeatures = data_xgbfeatures.fillna(0)

    # fit UMAP
    embedding = fit_umap(data_xgbfeatures)
    embedding.shape    

    # get samples incorrectly predicted 
    incorrect_preds = pd.read_csv('../figs_xgboost_validation/incorrectly_predicted_Evelonn_samples_model_' + str(model_num) + '.csv')
    
    to_highlight = Evelonn_diagnoses['accession'].isin(incorrect_preds['colname'])
    
    # make text labels - the prediction for incorrect predictions
    incorrect_preds.index = incorrect_preds['colname']
    text_labels = list(Evelonn_diagnoses['accession'].apply(lambda a: str(incorrect_preds.loc[a]['predictions']) if a in list(incorrect_preds['colname']) else ''))
    label_mapping = {'': '', '1': 'ccRCC', '2': 'chRCC', '3': 'onc', '4': 'pRCC'}
    text_labels = [label_mapping[t] for t in text_labels]

    # plot umap
    plot_umap_with_highlight(embedding, Evelonn_diagnoses['overallpath'], 'external_Evelonn_XGBoost_features' + str(model_num), list((['External Evelonn'] * len(Evelonn_diagnoses['overallpath']))), to_highlight, to_write = text_labels, s = 100)
    
```
